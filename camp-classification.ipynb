```python
## Train the model

print 'Training the model \n'

descriptorlist = []
designationlist=[]
poscnt=0
negcnt=0
ppc=(8, 8) # pixels per cell for HoG
path = 'camp_data/camp_train/'
for root, dirs, files in os.walk(path, topdown=False):
    for dname in dirs:
        if ('.pos' not in dname and '.neg' not in dname):
                continue
        des='camp' if ".pos" in dname else 'non-camp'
        for infile in os.listdir(path + dname):
            poscnt=poscnt + int(des=='camp') 
            negcnt=negcnt + int(des=='non-camp')
 #           print("current image is: " + infile + ", designation is:" + des)
            #coloredimagepil = Image.open(path + dname + "/" + infile)
            #coloredimage=np.array(coloredimagepil.getdata()).reshape(coloredimagepil.size[0], coloredimagepil.size[1], 3)
            coloredimage=io.imread(path + dname + "/" + infile)
            greyscale=skimage.color.rgb2grey(coloredimage)
            #greyscale=coloredimage.convert('1')

            edges = feature.canny(greyscale,  low_threshold=0.1, sigma=2.5)



    #HoG features computation
            fd, hog_image = hog(edges, orientations=8, pixels_per_cell=ppc,
                        cells_per_block=(1, 1), visualise=True)
            descriptorlist.append(fd)
            designationlist.append(des)

print 'training camp instances: ', poscnt, ', training non-camp instances: ', negcnt, '\n'
descriptorarray = np.dstack(descriptorlist)
descriptorarray=np.rollaxis(descriptorarray,-1)
trainingDescriptors=np.squeeze(descriptorarray)
designationarray = np.dstack(designationlist)
trainingDesignations=np.squeeze(designationarray)


## Create a classifier: a support vector classifier trained on training
## data of the set of camp and non-camp images
classifier = svm.SVC(gamma=0.001,class_weight='auto')
classifier.fit(trainingDescriptors, trainingDesignations) 


#classifier =  svm.LinearSVC()
#from sklearn.ensemble import RandomForestClassifier
#classifier = RandomForestClassifier(n_estimators=10)
#classifier = classifier.fit(trainingDescriptors, trainingDesignations)
```

```python
## Validate the model

print 'Validating the model \n'

descriptorlist = []
designationlist=[]

path = 'camp_data/camp_val/'
poscnt=0
negcnt=0
for root, dirs, files in os.walk(path, topdown=False):
    for dname in dirs:
        if ('.pos' not in dname and '.neg' not in dname):
                continue
        des='camp' if ".pos" in dname else 'non-camp'
        for infile in os.listdir(path + dname):
            poscnt=poscnt + int(des=='camp') 
            negcnt=negcnt + int(des=='non-camp')
#            print("current image is: " + infile + ", designation is:" + des)
            coloredimage = io.imread(path + dname + "/" + infile)
            greyscale=skimage.color.rgb2grey(coloredimage)
            edges = feature.canny(greyscale,  low_threshold=0.1, sigma=2.5)



    #HoG features computation
            fd, hog_image = hog(edges, orientations=8, pixels_per_cell=ppc,
                    cells_per_block=(1, 1), visualise=True)
            descriptorlist.append(fd)
            designationlist.append(des)


print 'validation camp instances: ', poscnt, ', validation non-camp instances: ', negcnt, '\n'
descriptorarray = np.dstack(descriptorlist)
descriptorarray=np.rollaxis(descriptorarray,-1)
validationDescriptors=np.squeeze(descriptorarray)
designationarray = np.dstack(designationlist)
validationDesignations=np.squeeze(designationarray)
     

## Now predict the image type on the validation set
predicted = classifier.predict(validationDescriptors)
print 'accuracy: ',classifier.score(validationDescriptors,validationDesignations), '\n' 
print 'confusion matrix: \n'
print metrics.confusion_matrix(validationDesignations,predicted)


print("expected")
print validationDesignations
print("predicted")
print predicted


    
```
